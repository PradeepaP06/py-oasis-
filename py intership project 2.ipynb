PROJECT 2 : WEATHER APP

```python
import requests
import tkinter as tk
from tkinter import messagebox

# --------------------------
# Command-Line Weather App
# --------------------------
def cli_weather_app():
    city = input("Enter the city name: ")
    api_key = "YOUR_API_KEY"  # Replace with your OpenWeatherMap API key
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

    try:
        response = requests.get(url)
        data = response.json()

        if data["cod"] == 200:
            temperature = data["main"]["temp"]
            humidity = data["main"]["humidity"]
            condition = data["weather"][0]["description"].title()

            print("\n----------------------------------------")
            print(f"Weather Report for: {city.title()}")
            print("----------------------------------------")
            print(f"Temperature: {temperature}°C")
            print(f"Humidity: {humidity}%")
            print(f"Condition: {condition}")
            print("----------------------------------------")
        else:
            print("\nCity not found. Please check the spelling and try again.")

    except Exception as e:
        print("\nError fetching data. Please check your internet connection.")


# --------------------------
# GUI Weather App
# --------------------------
def gui_weather_app():
    def get_weather():
        city = city_entry.get()
        api_key = "YOUR_API_KEY"  # Replace with your OpenWeatherMap API key
        url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

        try:
            response = requests.get(url)
            data = response.json()

            if data["cod"] == 200:
                temperature = data["main"]["temp"]
                humidity = data["main"]["humidity"]
                condition = data["weather"][0]["description"].title()

                result_label.config(
                    text=f"City: {city.title()}\nTemperature: {temperature}°C\nHumidity: {humidity}%\nCondition: {condition}"
                )
            else:
                messagebox.showerror("Error", "City not found. Please try again.")
        except:
            messagebox.showerror("Error", "Network error. Please check your connection.")

    window = tk.Tk()
    window.title("Weather Application")
    window.geometry("300x250")
    window.config(bg="#dce6f1")

    tk.Label(window, text="Enter City Name:", bg="#dce6f1", font=("Arial", 12, "bold")).pack(pady=10)
    city_entry = tk.Entry(window, width=25)
    city_entry.pack(pady=5)

    tk.Button(window, text="Get Weather", command=get_weather, bg="#4a90e2", fg="white", font=("Arial", 10, "bold")).pack(pady=10)
    result_label = tk.Label(window, text="", bg="#dce6f1", font=("Arial", 11))
    result_label.pack(pady=10)

    window.mainloop()


# --------------------------
# Choose Mode
# --------------------------
choice = input("Choose mode: (1) Command-Line or (2) GUI: ")

if choice == "1":
    cli_weather_app()
else:
    gui_weather_app()
```
