PROJECT 3 : CHAT APPLICATION

```python
import socket, threading, time

def server():
    s = socket.socket(); s.bind(('localhost', 12345)); s.listen(1)
    print("Server waiting...")
    conn, _ = s.accept(); print("Client connected!")
    while True:
        msg = conn.recv(1024).decode()
        if not msg or msg.lower() == 'bye': break
        print("Client:", msg)
        conn.send(input("You: ").encode())
    conn.close()

def client():
    time.sleep(1)  # Wait for server to start
    c = socket.socket(); c.connect(('localhost', 12345))
    while True:
        c.send(input("You: ").encode())
        msg = c.recv(1024).decode()
        if not msg or msg.lower() == 'bye': break
        print("Server:", msg)
    c.close()

threading.Thread(target=server, daemon=True).start()
client()
```
